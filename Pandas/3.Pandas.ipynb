### iloc

            int based location

            indexing :
                            df.iloc[row_index, col_index]
            slicing  :
                            df.iloc[row_slicing, col_slicing]

### loc

            label based location

            indexing :
                            df.iloc[row_label, col_label]
            slicing  :
                            df.iloc[row_label_slicing, col_label_slicing]
