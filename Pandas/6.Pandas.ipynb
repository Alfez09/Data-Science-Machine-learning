```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'gender':pd.Categorical(np.random.choice(['male', 'female'], 20, True)),
    'country':np.random.choice(['india', 'usa', 'china', 'aus'], 20, True),
    'salary':np.random.randint(30, 120, 20),
    'height':np.random.randint(120, 190, 20),
    'weight':np.random.randint(50, 90, 20)
}, index=[chr(65+i) for i in range(20)])
```

```python
df
```

```python
d = df.groupby(['country', 'gender']).agg({
    'salary':['mean'],
    'height':['mean', 'median'],
    'weight':['median']
})
d
```
