### pivot table

```python
df
```

```python
df.pivot_table(values=['salary', 'height', 'weight'], index='country')
```

```python
df.pivot_table(values=['salary', 'height', 'weight'], index='country', aggfunc='median')
```

```python
df.pivot_table(values=['salary', 'height', 'weight'], index='country', columns='gender', aggfunc='median')
```

```python
df.pivot_table(values=['salary', 'height', 'weight'], index=['country', 'gender'], aggfunc='median')
```
