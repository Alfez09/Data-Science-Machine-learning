### normal distribution

```python
normal = np.random.normal(140, 20, 1000)
print(normal)
```

```python
plt.figure(dpi=120)
plt.hist(normal, ec='k', bins=30)
plt.xlabel('Data', size=15)
plt.ylabel('Freq', size=15)
plt.title('Normal Distribution', size=15)
plt.show()
```

```python
mean = np.mean(normal)
mean
```

```python
std = np.std(normal)
std
```

```python
plt.figure(dpi=120)
plt.hist(normal, ec='k', bins=30 )
plt.plot([mean, mean], [0, 100], label='Mean')
plt.plot([mean+std, mean+std], [0, 100], label='Mean+std')
plt.plot([mean-std, mean-std], [0, 100], label='Mean-std')

plt.xlabel('Data', size=15)
plt.ylabel('Freq', size=15)
plt.title('Normal Distribution', size=15)
plt.show()
```

__In Normal distribution  std cover majority of data points / observation__

```python
plt.figure(dpi=120)
plt.hist(normal, ec='k', bins=30)
plt.plot([mean, mean], [0, 100], label='Mean')
plt.plot([mean+2*std, mean+2*std], [0, 100], label='Mean+2*std')
plt.plot([mean-2*std, mean-2*std], [0, 100], label='Mean-2*std')

plt.xlabel('Data', size=15)
plt.ylabel('Freq', size=15)
plt.title('Normal Distribution', size=15)
plt.show()
```

### Uniform distribution

```python
uniform = np.random.uniform(120, 190, 1000)
print(uniform)
```

```python
plt.figure(dpi=120)
plt.hist(uniform, ec='y')
plt.show()
```

```python
plt.figure(figsize=(13,6), dpi=120)
plt.hist(uniform, ec='y', bins=30)
mean = np.mean(uniform)
dev = np.std(uniform)
plt.title('Uniform Distribution', size=20)
plt.xlabel('Data', size=20)
plt.ylabel('Frequency', size=20)
plt.plot([mean,mean], [0, 60], lw=2)
plt.plot([mean-std, mean-std], [0, 60], lw=2)
plt.plot([mean+std, mean+std], [0, 60], lw=2)
plt.show()
```
