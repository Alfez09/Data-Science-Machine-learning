```python
def mode(arr):
    d = dict()
    for i in arr:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    ls = sorted(d.items(), key=lambda x: x[1], reverse=True)
    return ls[0][0]
    
```

#### Median For Odd number of values 

$$ i = \frac{N+1} {2} $$

$$ median = arr[i-1] $$

#### Median For Even number of values 

$$ i_1 = \frac{N} {2} $$

$$ i_2 = \frac{N+2} {2} $$

$$ median = \frac{arr[i_1-1] + arr[i_2 - 1]} {2} $$

```python
def median(arr):
    arr.sort()
    if len(arr) % 2 == 0:
        i1 = len(arr)//2
        i2 = (len(arr) + 2) // 2
        med = (arr[i1-1] + arr[i2-1]) / 2
        return med
    else:
        i = (len(arr) + 1) // 2
        med = arr[i - 1]
        return med
```

```python
data = list(np.random.randint(20, 30, 101))
print(data)
```

```python
np.median(data)
```

```python
median(data)
```

```python
class Stats:
    def __init__(self, arr):
        self.length = len(arr)

    def mean(self, arr):
        return sum(arr) / self.length

    def median(self, arr):
        arr.sort()
        if self.length % 2 == 0:
            i1 = self.length//2
            i2 = (self.length + 2) // 2
            med = (arr[i1-1] + arr[i2-1]) / 2
            return med
        else:
            i = (self.length + 1) // 2
            med = arr[i - 1]
            return med

    def mode(self, arr):
        d = dict()
        for i in arr:
            if i not in d:
                d[i] = 1
            else:
                d[i] += 1
        ls = sorted(d.items(), key=lambda x: x[1], reverse=True)
        return ls[0][0]
        
```

```python
a = Stats(data)
```

```python
a.mean(data)
```

```python
a.median(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heights = list(np.random.randint(120, 190, 100))
print(heights)
```

```python
plt.plot(heights)
plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=120)
plt.plot(heights, 'or')
plt.xlabel('Index')
plt.ylabel('Heights')
plt.title('Student height data')
plt.show()
```

```python
mean = np.mean(heights)
median = np.median(heights)
print(mean)
print(median)
```

```python
plt.figure(figsize=(12, 6), dpi=120)
plt.plot(heights, 'or')
plt.plot([0, 100], [mean, mean])
plt.xlabel('Index')
plt.ylabel('Heights')
plt.title('Student height data')
plt.show()
```

```python
plt.figure(figsize=(12, 6), dpi=120)
plt.plot(heights, 'or')
plt.plot([0, 100], [mean, mean], label='Mean')
plt.plot([0, 100], [median, median], label='Median')
plt.xlabel('Index', size=15)
plt.ylabel('Heights', size=15)
plt.title('Mean & median without Outlier', size=15)
plt.legend()
plt.show()
```
