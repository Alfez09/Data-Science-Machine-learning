```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heights = list(np.random.randint(120, 190, 100))
print(heights)
```

```python
mean = np.mean(heights)
median = np.median(heights)
```

```python
mean
```

```python
median
```

```python
plt.figure(figsize=(12, 6), dpi=120)
plt.plot(heights, 'or')
plt.plot([0, 100], [mean, mean], label='Mean')
plt.plot([0, 100], [median, median], label='Median')
plt.xlabel('Index', size=15)
plt.ylabel('Heights', size=15)
plt.title('Mean & median without Outlier', size=15)
plt.legend()
plt.show()
```

## Measures of Spread

    1. Deviation
    2. Standard Deviation
    3. Variance

## deviation
                    
                    dev = mean - data

```python
dev = [(mean - i) for i in heights]
print(dev)
```

```python
sum(dev) / len(dev)
```

        This tells us that the points are not getting deviate from mean 
    

```python
dev = [abs(mean - i) for i in heights]
print(dev)
```

```python
sum(dev) / len(dev)
```

```python
std = np.std(heights)
std
```

### Drawback of deviation
        Deviation only covers the nearby values. So the large values
        get neglected
        
        So we will overcome this drawback using variance

```python
plt.figure(figsize=(12, 6), dpi=120)
plt.plot(heights, 'or')
plt.plot([0, 100], [mean, mean], label='Mean')
plt.plot([0, 100], [mean+std, mean+std], label="Mean+Std")
plt.plot([0, 100], [mean-std, mean-std], label="Mean-Std")
plt.xlabel('Index', size=15)
plt.ylabel('Heights', size=15)
plt.title('Mean & median without Outlier', size=15)
plt.legend()
plt.show()
```

### Variance

$$ variance = \frac{\sum\limits_{i=1}^N (x_i - \bar x)^2} {N} $$

```python
var = [(mean - i)**2 for i in heights]
print(var)
```

```python
var = np.var(heights)
var
```

```python
plt.figure(figsize=(12, 6), dpi=120)
plt.plot(heights, 'or')
plt.plot([0, 100], [mean, mean], label='Mean')
plt.plot([0, 100], [mean+std, mean+std], label="Mean+Std")
plt.plot([0, 100], [mean-std, mean-std], label="Mean-Std")
plt.plot([0, 100], [mean+var, mean+var], label="Mean+var")
plt.plot([0, 100], [mean-var, mean-var], label="Mean-var")
plt.xlabel('Index', size=15)
plt.ylabel('Heights', size=15)
plt.title('Mean & median without Outlier', size=15)
plt.legend()
plt.show()
```
