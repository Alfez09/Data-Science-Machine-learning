## Quantile

    Arr --> q1, q2, q3
    
        q2 --> median
        q1 --> median of values less than Q2
        q3 --> median of values greater than q2

```python
normal = list(np.random.normal(140, 20, 1000))
print(normal)
```

```python
plt.figure(dpi=120)
plt.hist(normal, bins=30, ec='y')
plt.show()
```

```python
mean = np.mean(normal)
median = np.median(normal)
std = np.std(normal)
quantile = np.quantile(normal, [.25, .50, .75])
```

    Skewness:
                    mean == median    no skewness

                    mean < median     left skewness

                    mean > median     right skewness
